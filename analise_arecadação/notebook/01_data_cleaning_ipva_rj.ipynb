# # Análise de arrecadação de IPVA no estado do Rio de Janeiro


Este projeto tem como objetivo analisar a arrecadação tributária no estado do Rio de Janeiro, considerando valores arrecadados por município e por região administrativa.

A análise utiliza uma série histórica dos últimos cinco anos de arrecadação do seguinte tributo estadual:

- IPVA (Imposto sobre a Propriedade de Veículos Automotores)

Os dados são disponibilizados publicamente pelo portal de dados abertos do governo brasileiro, reunindo informações de arrecadação tributária divulgadas pela Secretaria de Fazenda do Estado do Rio de Janeiro.

Fonte dos dados: dados.gov.br  
Link direto: https://dados.gov.br/dados/conjuntos-dados/arrecadacao-receitas  

Última atualização do dataset: 28/07/2025  

Fase atual da análise: **Tratamento de dados**

```python
# Importações
import pandas as pd
```

```python
df = pd.read_csv(
    "../data/00_03_arrecadacao_por_municipio_e_regiao_ipva-2017.csv",
    sep=";",
    skiprows=2,          # ignora as linhas de título e ";;;;;;;;"
    encoding="latin1",   # funciona melhor para arquivos brasileiros antigos
    engine="python",
    decimal=",",
    thousands=".",
    index_col=False
)
```

## Conclusão do Processo de Tratamento de Dados

Este notebook teve como objetivo realizar o **tratamento e preparação dos dados de arrecadação de IPVA do estado do Rio de Janeiro referentes ao ano de 2017**, com foco em transformar um arquivo bruto em um **dataset limpo, consistente e pronto para análise**.

Durante o processo foram identificados diversos problemas estruturais no arquivo original, o que exigiu uma abordagem investigativa e incremental. Entre os principais desafios encontrados estão:

* **Problemas de codificação (encoding)** que resultaram em caracteres corrompidos, como “ï¿½”, afetando nomes de regiões e municípios.
* **Cabeçalhos inconsistentes**, incluindo colunas com nomes incorretos ou unificados de forma indevida (ex.: meses combinados como “Março/Abril”).
* **Formatação inadequada dos valores numéricos**, que utilizavam separadores de milhar e vírgula decimal no padrão brasileiro, dificultando a conversão para valores numéricos utilizáveis em análise.
* **Presença de linhas agregadas**, como “Total Geral”, que não representam observações individuais e poderiam distorcer análises estatísticas.
* **Duplicação conceitual**, como a coexistência das entradas “CAPITAL” e “RIO DE JANEIRO”, que representavam a mesma entidade.
* **Mistura de níveis de agregação**, pois o dataset apresenta tanto **municípios individuais** quanto **regiões administrativas**, sendo estas últimas compostas pela soma de vários municípios.

Ao longo do notebook foram aplicadas diversas técnicas de limpeza e organização de dados, incluindo:

* correção de codificação e padronização de texto;
* ajuste e renomeação de colunas;
* separação e reorganização de campos;
* conversão de valores monetários para formato numérico adequado;
* remoção de registros agregados ou redundantes;
* verificação de valores nulos, duplicidades e tipos de dados;
* validação estatística básica do dataset.

O processo também evidenciou a importância de **exploração e validação contínua dos dados**, utilizando ferramentas como ordenação de valores, inspeção de tipos (`dtypes`), estatísticas descritivas e análise de unicidade.

Como resultado final, foi produzido um **dataset estruturado e confiável**, contendo registros mensais de arrecadação de IPVA por região ou município, com valores numéricos padronizados e sem inconsistências estruturais relevantes.

Este conjunto de dados tratado será salvo como uma **versão final limpa**, que servirá de base para a próxima etapa do projeto:
a **análise exploratória e geração de visualizações**, a serem desenvolvidas em um novo notebook dedicado à interpretação dos dados, construção de gráficos e elaboração de relatórios analíticos.

Dessa forma, este notebook cumpre o papel fundamental de **etapa de preparação de dados (Data Cleaning / Data Preparation)** dentro do fluxo de trabalho de análise de dados, garantindo que as análises posteriores sejam realizadas sobre informações confiáveis e bem estruturadas.
